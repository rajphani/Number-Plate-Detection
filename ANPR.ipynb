```
image = cv2.imread('69590d1226627898t-high-security-registration-plates-hsrp-india-d1.jpg') 
```

```
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
median = cv2.medianBlur(gray,9)
```

```
from matplotlib import pyplot as plt

plt.imshow(median, cmap = 'gray', interpolation = 'bicubic')
plt.xticks([]), plt.yticks([])  # to hide tick values on X and Y axis
plt.show()
```

```
sobelx = cv2.Sobel(median,cv2.CV_64F,1,0,ksize=5)
#sobelxy = cv2.Sobel(sobelx,cv2.CV_64F,0,1,ksize=5)
#sobely = cv2.Sobel(median,cv2.CV_64F,0,1,ksize=5)
edges = cv2.Canny(median,100,200)
```
